```python
import math
def calculate_frobenius_norm(matrix):
  sum_of_squares = 0
  for row in matrix:
    for element in row:
      sum_of_squares += element**2

  forbenius_norm = math.sqrt(sum_of_squares)
  return forbenius_norm
```

```python
#Implement a 2D convolution manually (no np.convolve, no loops).
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view

def convolve2d_manual(image, kernel):
    kernel = np.flip(kernel, axis=(0, 1))
    patches = sliding_window_view(image, kernel.shape)
    return np.einsum('ijkl,kl->ij', patches, kernel)
```

```python
image = np.array([
    [1, 2, 3, 0],
    [0, 1, 2, 3],
    [3, 0, 1, 2],
    [2, 3, 0, 1]
])

kernel = np.array([
    [0, 1],
    [2, 3]
])

print(convolve2d_manual(image, kernel))
```

```python
#Write a function to compute softmax using only NumPy
import numpy as np

def softmax(x):
    x_shifted = x - np.max(x, axis=-1, keepdims=True)
    exp_x = np.exp(x_shifted)
    return exp_x / np.sum(exp_x, axis=-1, keepdims=True)
```

```python
#Implement a sigmoid function and its derivatives.
def sigmoid_function(x):
  return 1/(1+np.exp(-x))
def sigmoid_derivative(x):
   s = sigmoid_function(x)
   return s*(1-s)
```
