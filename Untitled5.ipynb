```python
# -------------------------------------------------------------
# QUESTION 1 (from x): DATA LOADING AND INITIAL EXPLORATION
# -------------------------------------------------------------
# Objective:
#    Create a sample time-series dataset (10 years of monthly sales)
#    Load into Pandas DataFrame
#    Display head(), info(), describe()
#    Plot the time series
# -------------------------------------------------------------

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Set a consistent plot style
sns.set_style("whitegrid")

# ---- 1. Create sample time-series dataset ----
np.random.seed(42)
dates = pd.date_range(start='2010-01-01', periods=120, freq='MS')   # 10 years of monthly data
data = 50 + np.arange(120)*0.5 + np.sin(np.arange(120)*0.5)*20 + np.random.randn(120)*10

df = pd.DataFrame({'Date': dates, 'Sales': data})
df.set_index('Date', inplace=True)

# ---- 2. Display basic dataset info ----
print("--- Data Head ---")
print(df.head())

print("\n--- Data Info ---")
df.info()

print("\n--- Data Description ---")
print(df.describe())

# ---- 3. Plot the time series ----
plt.figure(figsize=(12, 6))
sns.lineplot(data=df, x=df.index, y='Sales')
plt.title('Monthly Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```

```python
# -------------------------------------------------------------
# QUESTION 2 (from x): TIME SERIES DECOMPOSITION
# -------------------------------------------------------------
# Objective:
#    Decompose the time series into:
#         - Trend
#         - Seasonality
#         - Residual
#    Visualize the decomposed components
#
# Notes:
#    Using statsmodels.seasonal_decompose
#    'period=12' since monthly data  12 months = 1 season
# -------------------------------------------------------------

from statsmodels.tsa.seasonal import seasonal_decompose
import matplotlib.pyplot as plt

# ---- 1. Perform additive decomposition ----
decomposition = seasonal_decompose(df['Sales'], model='additive', period=12)

# ---- 2. Plot the decomposed components ----
fig = decomposition.plot()
fig.set_size_inches(10, 8)
plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# QUESTION 3 (from x): STATIONARITY TEST (ADF TEST)
# -------------------------------------------------------------
# Objective:
#    Test if the time series is stationary using the
#       Augmented Dickey-Fuller (ADF) test.
#    If non-stationary, apply differencing and test again.
#    Plot the differenced series.
# -------------------------------------------------------------

from statsmodels.tsa.stattools import adfuller
import seaborn as sns
import matplotlib.pyplot as plt

# ---- 1. ADF Test on the original series ----
print("\n--- ADF Test for Original Series ---")
result = adfuller(df['Sales'])

print(f'ADF Statistic: {result[0]:.4f}')
print(f'p-value: {result[1]:.4f}')
print('Critical Values:')
for key, value in result[4].items():
    print(f'   {key}: {value:.4f}')

if result[1] < 0.05:
    print("Interpretation: The original series is STATIONARY.")
else:
    print("Interpretation: The original series is NON-stationary. Differencing needed.")

# ---- 2. Differencing the series ----
df['Sales_Diff'] = df['Sales'].diff()

# ---- 3. ADF Test on the differenced series ----
print("\n--- ADF Test for Differenced Series ---")
result_diff = adfuller(df['Sales_Diff'].dropna())

print(f'ADF Statistic: {result_diff[0]:.4f}')
print(f'p-value: {result_diff[1]:.4f}')
print('Critical Values:')
for key, value in result_diff[4].items():
    print(f'   {key}: {value:.4f}')

if result_diff[1] < 0.05:
    print("Interpretation: The differenced series is STATIONARY.")
else:
    print("Interpretation: The differenced series is still NON-stationary.")

# ---- 4. Plot the differenced series ----
plt.figure(figsize=(12, 4))
sns.lineplot(x=df.index, y=df['Sales_Diff'])
plt.title('Differenced Monthly Sales')
plt.xlabel('Date')
plt.ylabel('Sales Difference')
plt.grid(True)
plt.show()
```

```python
# -------------------------------------------------------------
# QUESTION 4 (from x): ACF & PACF PLOTS
# -------------------------------------------------------------
# Objective:
#    Plot the Autocorrelation Function (ACF)
#    Plot the Partial Autocorrelation Function (PACF)
#    Use these plots to help identify ARIMA model orders:
#         - ACF   suggests q (MA order)
#         - PACF  suggests p (AR order)
# -------------------------------------------------------------

from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
import matplotlib.pyplot as plt

# ---- 1. Plot ACF and PACF for the differenced series ----
plt.figure(figsize=(14, 6))

# ACF plot
plt.subplot(1, 2, 1)
plot_acf(df['Sales_Diff'].dropna(), lags=30, ax=plt.gca())
plt.title('Autocorrelation Function (ACF) - Differenced Sales')

# PACF plot
plt.subplot(1, 2, 2)
plot_pacf(df['Sales_Diff'].dropna(), lags=30, ax=plt.gca())
plt.title('Partial Autocorrelation Function (PACF) - Differenced Sales')

plt.tight_layout()
plt.show()
```

```python
# -------------------------------------------------------------
# QUESTION 5 (from x): ARIMA MODEL
# -------------------------------------------------------------
# Objective:
#    Fit an ARIMA(p, d, q) model on the time series.
#    Use ACF/PACF + stationarity test to pick p, d, q.
#    Train on 80% of data, test on the remaining 20%.
#    Make predictions and compute RMSE.
#    Plot actual vs predicted values.
#
# Note:
#    Here we demonstrate ARIMA(1,1,1).
#    Replace (1,1,1) with your chosen parameters.
# -------------------------------------------------------------

from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt

# ---- 1. Train-test split ----
train_size = int(len(df) * 0.8)
train = df['Sales'][:train_size]
test  = df['Sales'][train_size:]

# ---- 2. Fit ARIMA model ----
order = (1, 1, 1)   # Example ARIMA order (p,d,q)
model_arima = ARIMA(train, order=order)
model_arima_fit = model_arima.fit()

print("\n--- ARIMA Model Summary ---")
print(model_arima_fit.summary())

# ---- 3. Make predictions ----
start_index = len(train)
end_index   = len(df) - 1

predictions_arima = model_arima_fit.predict(
    start=start_index,
    end=end_index,
    typ='levels'
)

# ---- 4. Evaluate using RMSE ----
rmse_arima = sqrt(mean_squared_error(test, predictions_arima))
print(f"\nRMSE for ARIMA{order}: {rmse_arima:.4f}")

# ---- 5. Plot actual vs predicted ----
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Actual Test Data')
plt.plot(predictions_arima.index, predictions_arima, color='red', label='ARIMA Predictions')

plt.title(f'ARIMA{order} Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -------------------------------------------------------------
# QUESTION 6 (from x): SARIMA MODEL
# -------------------------------------------------------------
# Objective:
#    Fit a SARIMA(p, d, q)(P, D, Q, s) model on the time series.
#    Handle seasonal patterns (e.g., monthly  s = 12).
#    Train on 80% of the data and test on the remaining 20%.
#    Make predictions and compute RMSE.
#    Plot actual vs predicted values.
#
# Notes:
#    Example parameters:
#         Non-seasonal: (p, d, q) = (1, 1, 1)
#         Seasonal:      (P, D, Q, s) = (1, 1, 0, 12)
#    Replace with values selected using ACF/PACF + seasonal analysis.
# -------------------------------------------------------------

from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt

# ---- 1. Train-test split (already done in Q5, but included for standalone clarity) ----
train_size = int(len(df) * 0.8)
train = df['Sales'][:train_size]
test  = df['Sales'][train_size:]

# ---- 2. Fit SARIMA model ----
sarima_order = (1, 1, 1)           # Non-seasonal (p, d, q)
seasonal_order = (1, 1, 0, 12)     # Seasonal (P, D, Q, s)

model_sarima = SARIMAX(
    train,
    order=sarima_order,
    seasonal_order=seasonal_order
)

model_sarima_fit = model_sarima.fit(disp=False)

print("\n--- SARIMA Model Summary ---")
print(model_sarima_fit.summary())

# ---- 3. Make predictions ----
start_index = len(train)
end_index   = len(df) - 1

predictions_sarima = model_sarima_fit.predict(
    start=start_index,
    end=end_index,
    typ='levels'
)

# ---- 4. Evaluate using RMSE ----
rmse_sarima = sqrt(mean_squared_error(test, predictions_sarima))
print(f"\nRMSE for SARIMA{sarima_order}{seasonal_order}: {rmse_sarima:.4f}")

# ---- 5. Plot predictions vs actual ----
plt.figure(figsize=(12, 6))
plt.plot(train.index, train, label='Training Data')
plt.plot(test.index, test, label='Actual Test Data')
plt.plot(predictions_sarima.index, predictions_sarima, color='green', label='SARIMA Predictions')

plt.title(f'SARIMA{sarima_order}{seasonal_order} Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -------------------------------------------------------------
# EXPERIMENT 8: LSTM FOR TIME SERIES FORECASTING
# -------------------------------------------------------------

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error
from math import sqrt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense

# -------------------------------------------------------------
# 1. CREATE THE DATAFRAME (from Experiment 1)
# -------------------------------------------------------------
np.random.seed(42)
dates = pd.date_range(start='2010-01-01', periods=120, freq='MS')
data = 50 + np.arange(120)*0.5 + np.sin(np.arange(120)*0.5)*20 + np.random.randn(120)*10

df = pd.DataFrame({'Date': dates, 'Sales': data})
df.set_index('Date', inplace=True)

# -------------------------------------------------------------
# 2. SCALE THE DATA
# -------------------------------------------------------------
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(df['Sales'].values.reshape(-1, 1))

# -------------------------------------------------------------
# 3. CREATE SEQUENCES
# -------------------------------------------------------------
def create_sequences(data, seq_length):
    X, y = [], []
    for i in range(len(data) - seq_length):
        X.append(data[i : i + seq_length, 0])
        y.append(data[i + seq_length, 0])
    return np.array(X), np.array(y)

sequence_length = 12
X, y = create_sequences(scaled_data, sequence_length)

# -------------------------------------------------------------
# 4. TRAINTEST SPLIT
# -------------------------------------------------------------
train_size_lstm = int(len(X) * 0.8)
X_train, X_test = X[:train_size_lstm], X[train_size_lstm:]
y_train, y_test = y[:train_size_lstm], y[train_size_lstm:]

# -------------------------------------------------------------
# 5. RESHAPE FOR LSTM   (samples, time_steps, features)
# -------------------------------------------------------------
X_train = X_train.reshape((X_train.shape[0], X_train.shape[1], 1))
X_test  = X_test.reshape((X_test.shape[0],  X_test.shape[1],  1))

# -------------------------------------------------------------
# 6. BUILD LSTM MODEL
# -------------------------------------------------------------
model_lstm = Sequential()
model_lstm.add(LSTM(50, activation='relu', input_shape=(sequence_length, 1)))
model_lstm.add(Dense(1))
model_lstm.compile(optimizer='adam', loss='mse')

# -------------------------------------------------------------
# 7. TRAIN THE MODEL
# -------------------------------------------------------------
history = model_lstm.fit(X_train, y_train, epochs=100, batch_size=1, verbose=0)

# -------------------------------------------------------------
# 8. MAKE PREDICTIONS
# -------------------------------------------------------------
predictions_scaled = model_lstm.predict(X_test)

# Inverse transform
predictions_lstm = scaler.inverse_transform(predictions_scaled)
y_test_original   = scaler.inverse_transform(y_test.reshape(-1, 1))

# -------------------------------------------------------------
# 9. EVALUATE THE MODEL (RMSE)
# -------------------------------------------------------------
rmse_lstm = sqrt(mean_squared_error(y_test_original, predictions_lstm))
print(f"\nRMSE for LSTM (sequence_length = {sequence_length}): {rmse_lstm:.4f}")

# -------------------------------------------------------------
# 10. PLOT THE RESULTS
# -------------------------------------------------------------
plt.figure(figsize=(12, 6))

# Last len(y_test_original) dates correspond to test predictions
pred_idx = df.index[-len(y_test_original):]

plt.plot(pred_idx, y_test_original, label='Actual Test Data')
plt.plot(pred_idx, predictions_lstm, color='purple', label='LSTM Predictions')

plt.title('LSTM Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# -------------------------------------------------------------
# TASK 1 (from y): SIMPLE AGGREGATE & SIMPLE AVERAGE OF PRICE RELATIVES
# -------------------------------------------------------------
# Objective:
#    Create a DataFrame of Base Year (2020) and Current Year (2023) prices
#    Calculate:
#         1. Simple Aggregate Price Index
#         2. Price Relatives for each commodity
#         3. Simple Average of Price Relatives Index
#    Print and interpret results
# -------------------------------------------------------------

import pandas as pd

# ---- 1. Create the DataFrame ----
data = {
    'Commodity': ['A', 'B', 'C', 'D', 'E'],
    'Price_2020': [10, 15, 20, 25, 30],
    'Price_2023': [12, 18, 22, 28, 35]
}

df = pd.DataFrame(data)

# ---- 2. Calculate Simple Aggregate Price Index ----
simple_aggregate_index = (df['Price_2023'].sum() / df['Price_2020'].sum()) * 100

# ---- 3. Calculate Price Relatives ----
df['Price_Relative'] = (df['Price_2023'] / df['Price_2020']) * 100

# ---- 4. Simple Average of Price Relatives Index ----
simple_avg_price_relative_index = df['Price_Relative'].mean()

# ---- 5. Print Results ----
print(df)
print(f"\nSimple Aggregate Price Index: {simple_aggregate_index:.2f}")
print(f"Simple Average of Price Relatives Index: {simple_avg_price_relative_index:.2f}")

# ---- 6. Interpretation ----
print("\nInterpretation:")
print(" The Simple Aggregate Index compares total price levels in 2023 relative to 2020.")
print(" The Simple Average of Price Relatives treats all commodities equally, averaging individual price changes.")
print(" Values differ because the Simple Aggregate Index is influenced by absolute price levels,")
print("  while the simple average treats each commodity as equally important regardless of its price.")
```

```python
# -------------------------------------------------------------
# TASK 2 (from y): LASPEYRES, PAASCHE & FISHER'S PRICE INDICES
# -------------------------------------------------------------
# Objective:
#    Create a DataFrame with price & quantity data (2020 vs 2023)
#    Compute:
#         1. Laspeyres Price Index
#         2. Paasche Price Index
#         3. Fishers Ideal Price Index
# -------------------------------------------------------------

import pandas as pd

# ---- 1. Create DataFrame ----
data = {
    'Commodity': ['A', 'B', 'C'],
    'P_0': [5, 10, 15],     # Price in 2020
    'Q_0': [100, 50, 30],   # Quantity in 2020
    'P_1': [6, 12, 18],     # Price in 2023
    'Q_1': [90, 60, 35]     # Quantity in 2023
}

df2 = pd.DataFrame(data)

# ---- 2. Add all required product columns ----
df2['P0Q0'] = df2['P_0'] * df2['Q_0']
df2['P1Q0'] = df2['P_1'] * df2['Q_0']
df2['P0Q1'] = df2['P_0'] * df2['Q_1']
df2['P1Q1'] = df2['P_1'] * df2['Q_1']

# ---- 3. Calculate sums ----
sum_P0Q0 = df2['P0Q0'].sum()
sum_P1Q0 = df2['P1Q0'].sum()
sum_P0Q1 = df2['P0Q1'].sum()
sum_P1Q1 = df2['P1Q1'].sum()

# ---- 4. Laspeyres Price Index ----
laspeyres_index = (sum_P1Q0 / sum_P0Q0) * 100

# ---- 5. Paasche Price Index ----
paasche_index = (sum_P1Q1 / sum_P0Q1) * 100

# ---- 6. Fishers Ideal Price Index ----
fisher_index = (laspeyres_index * paasche_index) ** 0.5

# ---- 7. Print results ----
print(df2)

print(f"\nLaspeyres Price Index : {laspeyres_index:.2f}")
print(f"Paasche Price Index   : {paasche_index:.2f}")
print(f"Fisher's Ideal Index  : {fisher_index:.2f}")

# ---- 8. Interpretation ----
print("\nInterpretation:")
print(" Laspeyres uses base-year quantities (Q0)  generally overestimates inflation.")
print(" Paasche uses current-year quantities (Q1)  generally underestimates inflation.")
print(" Fisher's Index is the geometric mean of both  considered 'ideal' because")
print("  it balances the biases of Laspeyres and Paasche.")
```

```python
# -------------------------------------------------------------
# TASK 3 (from y): LASPEYRES, PAASCHE & FISHER'S QUANTITY INDICES
# -------------------------------------------------------------
# Objective:
#    Use the same pricequantity data from Task 2
#    Compute:
#         1. Laspeyres Quantity Index
#         2. Paasche Quantity Index
#         3. Fishers Ideal Quantity Index
#    These use price weights instead of quantity weights.
# -------------------------------------------------------------

import pandas as pd

# ---- 1. Recreate DataFrame (same as Task 2 for independent execution) ----
data = {
    'Commodity': ['A', 'B', 'C'],
    'P_0': [5, 10, 15],     # Price in 2020
    'Q_0': [100, 50, 30],   # Quantity in 2020
    'P_1': [6, 12, 18],     # Price in 2023
    'Q_1': [90, 60, 35]     # Quantity in 2023
}

df3 = pd.DataFrame(data)

# ---- 2. Compute required weighted sums ----
# Base-period price weights (Laspeyres Quantity Index)
sum_Q1P0 = (df3['Q_1'] * df3['P_0']).sum()
sum_Q0P0 = (df3['Q_0'] * df3['P_0']).sum()

# Current-period price weights (Paasche Quantity Index)
sum_Q1P1 = (df3['Q_1'] * df3['P_1']).sum()
sum_Q0P1 = (df3['Q_0'] * df3['P_1']).sum()

# ---- 3. Laspeyres Quantity Index ----
laspeyres_Q = (sum_Q1P0 / sum_Q0P0) * 100

# ---- 4. Paasche Quantity Index ----
paasche_Q = (sum_Q1P1 / sum_Q0P1) * 100

# ---- 5. Fishers Ideal Quantity Index ----
fisher_Q = (laspeyres_Q * paasche_Q) ** 0.5

# ---- 6. Print results ----

print(df3)

print(f"\nLaspeyres Quantity Index : {laspeyres_Q:.2f}")
print(f"Paasche Quantity Index   : {paasche_Q:.2f}")
print(f"Fisher's Quantity Index  : {fisher_Q:.2f}")

# ---- 7. Interpretation ----
print("\nInterpretation:")
print(" Laspeyres Quantity Index uses base-year prices (P0) as weights.")
print("   Measures how total quantity has changed using old price structure.")
print(" Paasche Quantity Index uses current-year prices (P1) as weights.")
print("   Measures quantity change under new price structure.")
print(" Fishers Quantity Index is the geometric mean of both and avoids")
print("  the upward bias of Laspeyres and downward bias of Paasche.")
```

```python
# -------------------------------------------------------------
# TASK 4 (from y): SHIFTING THE BASE YEAR OF AN INDEX
# -------------------------------------------------------------
# Objective:
#    Use a time series with base year 2010 = 100
#    Shift the base to 2016 = 100
#    Recalculate all index values
#    Plot original vs shifted index series
# -------------------------------------------------------------

import pandas as pd
import matplotlib.pyplot as plt

# ---- 1. Original index data ----
data = {
    'Year': [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020],
    'Original_Index': [100, 105, 112, 118, 125, 130, 138, 145, 152, 160, 168]
}

df4 = pd.DataFrame(data)

# ---- 2. Choose new base year: 2016 ----
new_base_year = 2016
base_value = df4.loc[df4['Year'] == new_base_year, 'Original_Index'].values[0]

# ---- 3. Calculate shifted index series ----
df4['Shifted_Index'] = (df4['Original_Index'] / base_value) * 100

# ---- 4. Print results ----
print(df4)

# ---- 5. Visualization ----
plt.figure(figsize=(12, 6))

plt.plot(df4['Year'], df4['Original_Index'], marker='o', label='Original Index (Base 2010 = 100)')
plt.plot(df4['Year'], df4['Shifted_Index'], marker='o', label='Shifted Index (Base 2016 = 100)')

# Mark the new base year
plt.axvline(new_base_year, color='red', linestyle='--', label='New Base Year (2016)')

plt.title('Original vs Shifted Index Series')
plt.xlabel('Year')
plt.ylabel('Index Value')
plt.legend()
plt.grid(True)
plt.show()

# ---- 6. Interpretation ----
print("\nInterpretation:")
print(" Shifting the base year re-scales the entire index series so that the chosen year (2016)")
print("  becomes 100 instead of 2010.")
print(" Relative movements stay the SAME  only the scale changes.")
print(" This makes comparison easier when analysts want a more recent base year.")
```

```python
# -------------------------------------------------------------
# TASK 5 (from y): COST OF LIVING INDEX (CLI)
# -------------------------------------------------------------
# Objective:
#    Compute the Cost of Living Index using the Aggregate Expenditure Method
#    Compare household expenditure in:
#         - Base Year: 2022 (P0, Q0)
#         - Current Year: 2024 (P1, Q0)
#    CLI = ( P1Q0 /  P0Q0) * 100
# -------------------------------------------------------------

import pandas as pd

# ---- 1. Create the DataFrame ----
data = {
    'Item': ['Food', 'Housing', 'Transport', 'Clothing', 'Entertainment'],
    'P_0': [50, 200, 30, 40, 25],   # Price in 2022
    'Q_0': [10, 1, 5, 2, 3],        # Quantity in 2022
    'P_1': [55, 220, 35, 42, 28]    # Price in 2024
}

df5 = pd.DataFrame(data)

# ---- 2. Base period expenditure: P0Q0 ----
df5['P0Q0'] = df5['P_0'] * df5['Q_0']

# ---- 3. Current period expenditure using base quantities: P1Q0 ----
df5['P1Q0'] = df5['P_1'] * df5['Q_0']

# ---- 4. Sum total expenditures ----
sum_P0Q0 = df5['P0Q0'].sum()
sum_P1Q0 = df5['P1Q0'].sum()

# ---- 5. Calculate Cost of Living Index ----
CLI = (sum_P1Q0 / sum_P0Q0) * 100

# ---- 6. Print results ----
print(df5)

print(f"\nTotal Expenditure in 2022 (Base Year): {sum_P0Q0}")
print(f"Total Expenditure in 2024 (Using Base Quantities): {sum_P1Q0}")

print(f"\nCost of Living Index (Base 2022 = 100): {CLI:.2f}")

# ---- 7. Interpretation ----
print("\nInterpretation:")
print(" The CLI shows how much the cost of purchasing the SAME basket of goods")
print("  (quantities from 2022) increased by 2024.")
print(" A CLI value above 100 means the cost of living has increased.")
print(f" A CLI of {CLI:.2f} means that the cost of living increased by approximately")
print(f"  {CLI - 100:.2f}% between 2022 and 2024.")
```

```python
import nltk
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer

# ---- Download required NLTK resources (run once) ----
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('punkt_tab') # Added to resolve LookupError

# ---- 1. Example text ----
text = "The quick brown fox jumps over the lazy dog. It was running very fast!"

# ---- 2. Sentence Tokenization ----
sentences = sent_tokenize(text)
print("Sentence Tokenization:")
print(sentences, "\n")

# ---- 3. Word Tokenization ----
words = word_tokenize(text)
print("Word Tokenization:")
print(words, "\n")

# ---- 4. Stopword Removal ----
stop_words = set(stopwords.words('english'))
filtered_words = [w for w in words if w.lower() not in stop_words and w.isalpha()]
print("After Stopword Removal:")
print(filtered_words, "\n")

# ---- 5. Stemming ----
ps = PorterStemmer()
stemmed_words = [ps.stem(w) for w in filtered_words]
print("Stemming Output:")
print(stemmed_words, "\n")

# ---- 6. Lemmatization ----
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(w) for w in filtered_words]
print("Lemmatization Output:")
print(lemmatized_words, "\n")

# ---- 7. Interpretation ----
print("Interpretation:")
print(" Tokenization breaks text into sentences and words.")
print(" Stopword removal eliminates common words with low meaning.")
print(" Stemming reduces words to crude roots (e.g., 'running'  'run').")
print(" Lemmatization reduces words to meaningful dictionary forms.")
```

```python
# -------------------------------------------------------------
# EXPERIMENT 2 (from z): CORPORA & LEXICAL RESOURCES
# -------------------------------------------------------------
# Objective:
#    Explore NLTK corpora (gutenberg, webtext, movie_reviews)
#    Perform basic corpus statistics:
#         - word count
#         - unique words
#         - lexical diversity
#    Explore WordNet:
#         - synsets
#         - hypernyms & hyponyms
#         - semantic similarity
# -------------------------------------------------------------

import nltk
from nltk.corpus import gutenberg, webtext, movie_reviews, wordnet as wn

# ---- Download necessary corpora (if not downloaded yet) ----
nltk.download('gutenberg')
nltk.download('webtext')
nltk.download('movie_reviews')
nltk.download('wordnet')

# -------------------------------------------------------------
# PART A  WORKING WITH NLTK CORPORA
# -------------------------------------------------------------

# ---- 1. Load a Gutenberg text ----
sample_text = gutenberg.raw("austen-emma.txt")

# ---- 2. Compute statistics ----
words = nltk.word_tokenize(sample_text)
unique_words = set(words)
lexical_diversity = len(unique_words) / len(words)

print("=== Gutenberg Corpus: 'Emma' by Jane Austen ===")
print(f"Total Words       : {len(words)}")
print(f"Unique Words      : {len(unique_words)}")
print(f"Lexical Diversity : {lexical_diversity:.4f}\n")

# ---- 3. Webtext example ----
web = webtext.raw('pirates.txt')
web_words = nltk.word_tokenize(web)
print("=== Webtext Corpus: Pirates ===")
print(f"Total Words: {len(web_words)}\n")

# ---- 4. Movie Reviews example ----
print("=== Movie Reviews Corpus ===")
print(f"Number of Positive Reviews : {len(movie_reviews.fileids('pos'))}")
print(f"Number of Negative Reviews : {len(movie_reviews.fileids('neg'))}\n")

# -------------------------------------------------------------
# PART B  WORDNET EXPLORATION
# -------------------------------------------------------------

# ---- 5. Synsets for a given word ----
word = "car"
car_synsets = wn.synsets(word)

print(f"=== WordNet Synsets for '{word}' ===")
for syn in car_synsets:
    print(f"{syn.name()} : {syn.definition()}")
print()

# ---- 6. Hypernyms & Hyponyms ----
car = wn.synset('car.n.01')

print("Hypernyms of 'car.n.01':")
print(car.hypernyms(), "\n")

print("Hyponyms of 'car.n.01':")
print(car.hyponyms(), "\n")

# ---- 7. Semantic Similarity ----
# Compare similarity between 'car' and 'bicycle'
bicycle = wn.synset('bicycle.n.01')

similarity = car.path_similarity(bicycle)
print("=== Semantic Similarity (path similarity) ===")
print(f"Car vs Bicycle Similarity: {similarity:.4f}\n")

# ---- 8. Interpretation ----
print("Interpretation:")
print(" Corpora let us analyze large text datasets and compute lexical statistics.")
print(" WordNet provides word meanings, synonyms, and relationships.")
print(" Hypernyms show general concepts (car  vehicle).")
print(" Hyponyms show specific types (car  sedan, hatchback).")
print(" Semantic similarity measures conceptual closeness between words.")
```

```python
# -------------------------------------------------------------
# EXPERIMENT 3 (from z): TEXT CLASSIFICATION
# -------------------------------------------------------------
# Objective:
#    Build a simple text classifier using the movie_reviews corpus
#    Steps:
#         1. Load pre-labeled dataset (positive / negative)
#         2. Extract features (Bag of Words)
#         3. Train Naive Bayes classifier
#         4. Evaluate accuracy on test data
# -------------------------------------------------------------

import nltk
from nltk.corpus import movie_reviews
import random

# ---- Download dataset ----
nltk.download('movie_reviews')

# -------------------------------------------------------------
# PART A  PREPARE DATA
# -------------------------------------------------------------

# ---- 1. Load documents with labels ----
documents = [
    (movie_reviews.words(fileid), category)
    for category in movie_reviews.categories()
    for fileid in movie_reviews.fileids(category)
]

# Shuffle for randomness
random.shuffle(documents)

# ---- 2. Build a list of all words ----
all_words = nltk.FreqDist(w.lower() for w in movie_reviews.words())

# Use the 3000 most common words as features
word_features = list(all_words)[:3000]

# ---- 3. Feature extractor ----
def document_features(document):
    words = set(document)
    features = {word: (word in words) for word in word_features}
    return features

# ---- 4. Create feature sets ----
feature_sets = [(document_features(doc), category) for (doc, category) in documents]

# -------------------------------------------------------------
# PART B  TRAIN-TEST SPLIT & MODEL TRAINING
# -------------------------------------------------------------

train_size = int(len(feature_sets) * 0.8)
train_set = feature_sets[:train_size]
test_set = feature_sets[train_size:]

# ---- 5. Train Naive Bayes Classifier ----
classifier = nltk.NaiveBayesClassifier.train(train_set)

# -------------------------------------------------------------
# PART C  EVALUATION
# -------------------------------------------------------------

accuracy = nltk.classify.accuracy(classifier, test_set)
print("=== Naive Bayes Classifier Accuracy ===")
print(f"Accuracy: {accuracy * 100:.2f}%\n")

print("Most Informative Features:")
classifier.show_most_informative_features(10)

# ---- 6. Interpretation ----
print("\nInterpretation:")
print(" The Bag-of-Words model represents each review as word presence/absence.")
print(" Naive Bayes assumes independence between words but performs well for text.")
print(" Accuracy indicates how well the classifier predicts positive/negative sentiment.")
print(" Informative features show words that strongly indicate sentiment.")
```

```python
# -------------------------------------------------------------
# EXPERIMENT 4 (from z): LINGUISTIC STRUCTURE ANALYSIS
# -------------------------------------------------------------
# Objective:
#    Analyze grammatical structure of text using:
#         1. POS Tagging
#         2. Named Entity Recognition (NER)
#         3. Chunking (phrase extraction)
#
# Example sentence:
#   "Barack Obama was born in Hawaii and became the President of the United States."
# -------------------------------------------------------------

import nltk
from nltk import pos_tag, word_tokenize, ne_chunk
from nltk.tree import Tree

# ---- Download required NLTK resources ----
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker_tab') # Added to fix the LookupError

# ---- 1. Example sentence ----
sentence = "Barack Obama was born in Hawaii and became the President of the United States."

# -------------------------------------------------------------
# PART A  PART-OF-SPEECH (POS) TAGGING
# -------------------------------------------------------------

tokens = word_tokenize(sentence)
pos_tags = pos_tag(tokens)

print("=== POS Tagging ===")
print(pos_tags)
print()

# -------------------------------------------------------------
# PART B  NAMED ENTITY RECOGNITION (NER)
# -------------------------------------------------------------

ner_tree = ne_chunk(pos_tags)

print("=== Named Entity Recognition ===")
print(ner_tree)
print()

# Extract and print named entities clearly
def extract_entities(tree):
    entities = []
    for subtree in tree:
        if isinstance(subtree, Tree):  # NER chunks are of type Tree
            entity = " ".join([token for token, pos in subtree.leaves()])
            entity_type = subtree.label()
            entities.append((entity, entity_type))
    return entities

entities = extract_entities(ner_tree)
print("Named Entities Found:", entities, "\n")

# -------------------------------------------------------------
# PART C  CHUNKING (NOUN PHRASE EXTRACTION)
# -------------------------------------------------------------

# Define a regex grammar for noun phrases (NP)
grammar = r"""
    NP: {<DT>?<JJ>*<NN.*>}   # determiner + adjectives + noun
        {<NNP>+}             # proper noun sequence
"""

chunk_parser = nltk.RegexpParser(grammar)
chunk_tree = chunk_parser.parse(pos_tags)

print("=== Chunking (Noun Phrases) ===")
print(chunk_tree)
print()

# -------------------------------------------------------------
# Interpretation
# -------------------------------------------------------------
print("Interpretation:")
print(" POS tagging assigns grammatical roles (noun, verb, adj, etc.) to each word.")
print(" NER identifies real-world entities such as PERSON, LOCATION, ORGANIZATION.")
print(" Chunking groups tokens into meaningful phrases such as noun phrases (NP).")
```
